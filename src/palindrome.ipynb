```python
import pandas as pd
```

```python
length = (10**9) # How much numbers we are going up to 
def isPalendrome(num): #check
    return (str(num)==str(num)[::-1])

def palindromeGenerator(num): # generating palendromes
    y = []
    for i in range(1, num + 1):
        if isPalendrome(i):  
            y.append(i) 
    return y
```

```python
palindromes = palindromeGenerator(length) #original palendrome
newPalendromes = []
counterArr= []
counter = 0
for i in range(1,len(palindromes)+1):
    if isPalendrome(i):
        counter = counter + 1
        newPalendromes.append(palindromes[i-1]) #left
        counterArr.append(counter) #right
```

```python
df = pd.DataFrame({
    "palindromic-index palindrome": newPalendromes,
    "ith/counter": counterArr,
}) 

print(df.to_string(index=False))

df.plot(x="ith/counter",y="palindromic-index palindrome")
```
