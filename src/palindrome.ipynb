```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
length = (10**6) # How much numbers we are going up to 
```

```python
def isPalendrome(num): #check
    return (str(num)==str(num)[::-1])
```

```python
def palindromeGenerator(num): # generating palendromes
    y = []
    for i in range(1, num + 1):
        if isPalendrome(i):  
            y.append(i) 
    return y
```

```python
palindromes = palindromeGenerator(length) #originalpalendrome
newPalendromes = []
counterArr= []
counter = 0
for i in range(1,len(palindromes)+1):
    if isPalendrome(i):
        counter = counter + 1
        newPalendromes.append(palindromes[i-1]) #left
        counterArr.append(counter) #right
        print(palindromes[i-1],counter,sep=" ") #the right side is the ith palendromic index palendrome
        # print(counter) # how many palindormic palindrome index
```

```python
import pandas as pd
import numpy as np  # Don't forget to import numpy

# Create DataFrame
df = pd.DataFrame({
    "palindromic-index palindrome": newPalendromes,
    "ith/counter": counterArr,
}) 

print(df.to_string(index=False))
```
